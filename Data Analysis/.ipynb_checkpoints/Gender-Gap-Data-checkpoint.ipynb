## Data Cleaning

```python
# Copy the original dataset
ggi_c = ggi.copy()
```

```python
ggi_c = ggi_c.drop(columns=['2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2020'])
ggi_c
```

```python
# Rename the column "Subindicator Type" and "Country Name"
ggi_c = ggi_c.rename(columns={"Subindicator Type": "SubindicatorType","Country Name":"Country"})
ggi_c
```

```python
# Due to our data exploration analysis (next section) - we found data Uzbekistan had no value. 
# We thereby decided to remove it from the dataset
uzb = ggi_c.loc[ggi_c.Country == "Uzbekistan"].index
uzb
```

```python
ggi_c.drop(uzb, axis=0, inplace=True)
```

```python
#reseting the index to fit the actual columns
ggi_c = ggi_c.reset_index()
ggi_c
```

## Data Exploration Analysis

### The Global Gender Gap Index (vis1)

```python
import matplotlib.pyplot as plt
```

```python
eda1 = ggi_c.copy()
```

```python
# Selecting all the rows in the column "Indicator", where the value is "Global Gender Gap Economic Participation and Opportunity Subindex"
eda1 = eda1.loc[eda1.Indicator == "Global Gender Gap Economic Participation and Opportunity Subindex"]
eda1
```

```python
# Selecting all the rows in the column where "SubindicatorType" is "Index":
eda1 = eda1.loc[eda1.SubindicatorType == "Index"]
eda1
```

```python
#Barchart representing the "Global Gender Gap in 'Economic Participation and Opportunity' in different countries for 2021"

plt.figure(figsize=(20, 20))

#Giving the visualization a title
plt.title("Global Gender Gap in 'Economic Participation and Opportunity' in different countries for 2021")

#Giving the x-axis a name
plt.xticks(rotation='vertical')
plt.xlabel("Countries")

#Giving the y-axis a name
plt.ylabel("Index")


plt.bar('Country','2021', data=eda1, width = 1, edgecolor=[1,1,1])
```

```python
# We divide the visualization into two parts to create a better overview
vis1 = eda1.iloc[0:78]
vis2 = eda1.iloc[78:]
```

```python
#Barchart Part 1
#Barchart representing the "Global Gender Gap in 'Economic Participation and Opportunity' in different countries for 2021"

plt.figure(figsize=(20, 20))

#Giving the visualization a title
plt.title("Global Gender Gap in 'Economic Participation and Opportunity' in different countries for 2021")

#Giving the x-axis a name
plt.xticks(rotation='vertical')
plt.xlabel("Countries")

#Giving the y-axis a name
plt.ylabel("Index")


plt.bar('Country','2021', data=vis1, width = 1, edgecolor=[1,1,1])
```

```python
#Barchart Part 1
#Barchart representing the "Global Gender Gap in 'Economic Participation and Opportunity' in different countries for 2021"

plt.figure(figsize=(20, 20))

#Giving the visualization a title
plt.title("Global Gender Gap in 'Economic Participation and Opportunity' in different countries for 2021")

#Giving the x-axis a name
plt.xticks(rotation='vertical')
plt.xlabel("Countries")

#Giving the y-axis a name
plt.ylabel("Index")


plt.bar('Country','2021', data=vis2, width = 1, edgecolor=[1,1,1])
```
